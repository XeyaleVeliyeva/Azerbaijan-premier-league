```python
positions = ['Goalkeeper', 'Centre-Back', 'Left-Back', 'Right-Back',
       'Central Midfield', 'Attacking Midfield', 'Left Winger',
       'Right Winger', 'Centre-Forward', 'Defensive Midfield',
       'Right Midfield', 'Left Midfield']

for i in positions:
  df["Futbolçunun adı"] = df["Futbolçunun adı"].str.replace(i," ")
```

```python
df
```

```python
# def currency(x):
#   if "€" in str(x):
#     return "EUR"
#   else:
#     return "Other"

# df["Pul vahidi"] = df["Dəyər"].apply(currency)
df["Pul vahidi"]="EUR"
```

```python
df["Dəyər"] = df["Dəyər"].str.replace("€","")
```

```python
num_zero_values = df[df["Dəyər"]=="€0"].shape[0]
print(f"Number of players with value '€0': {num_zero_values}")
```

```python
df["Dəyər"] = df["Dəyər"].astype(str)
m = df["Dəyər"].str.contains("m",na=False)
df.loc[m,"Dəyər"] = df.loc[m,"Dəyər"].str.replace("m","",regex=False).astype(float)*1_000_000

k = df["Dəyər"].str.contains("k",na=False)
df.loc[k,"Dəyər"] = df.loc[k,"Dəyər"].str.replace("k","",regex=False).astype(float)*1000

df["Dəyər"] = df["Dəyər"].astype(float).round().astype(int)
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.histplot(data = df,x="Yaş",kde=True)
plt.title("Futbolçuların yaşa görə paylanması",fontdict={"fontsize":16,"fontweight":"bold","color":"tab:blue"});
```

```python
value_position = df.groupby("Mövqe")["Dəyər"].mean().round().reset_index().head().sort_values(by="Dəyər",ascending=False)
value_position
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data =value_position,hue="Mövqe",y="Dəyər",palette="Blues_r" )
plt.title("Ən çox qazanan 5 mövqe",fontdict={"fontsize":16,"fontweight":"bold","color":"tab:blue"});
plt.xticks(rotation=45);
```
